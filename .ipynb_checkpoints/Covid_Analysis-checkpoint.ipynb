```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
import folium
from folium import plugins
import plotly.express as ex
import plotly.graph_objects as go
plt.rcParams['figure.figsize'] = 10,12
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("Covid cases in India.xlsx")
df.head()
```

```python
df_india = df.copy()
```

```python
# Coordinates of India States and Union Territories
ind_coor = pd.read_excel("Indian Coordinates.xlsx")
ind_coor.head()
```

```python
# Day by day data of India, Korea, Italy and Wuhan
dbd_India = pd.read_excel('per_day_cases.xlsx',parse_dates=True, sheet_name='India')
dbd_Italy = pd.read_excel('per_day_cases.xlsx',parse_dates=True, sheet_name="Italy")
dbd_Korea = pd.read_excel('per_day_cases.xlsx',parse_dates=True, sheet_name="Korea")
dbd_Wuhan = pd.read_excel('per_day_cases.xlsx',parse_dates=True, sheet_name="Wuhan")
```

## COVID19 cases in India

```python
df.drop(['S. No.'],axis=1,inplace=True)
```

```python
df['Total Cases'] = df['Total Confirmed cases (Indian National)']+ df['Total Confirmed cases ( Foreign National )']
df
```

```python
total_cases = df['Total Cases'].sum()
print("Total number of confirmed COVID cases across India (till 22nd March) :",total_cases)
```

```python
df.style.background_gradient(cmap="Blues")
```

### Visualization Inference
- Maharashtra has the highest number of confirmed cases of Indian National followed by Kerala.
- Haryana has the the highest number of confirmed cases of Foreign National.
- Maharashtra,Kerala,Delhi and U.P. had reported with maximum number of cases.

```python
# Total active cases in India
df['Total Active Cases'] = df['Total Cases'] - (df['Cured'] + df['Death'])
df
```

```python
total_active = df['Total Active Cases'].sum()
print("Total active cases across India :",total_active)
tot_cases = df.groupby("Name of State / UT")['Total Active Cases'].sum().sort_values(ascending = False).to_frame()
tot_cases.style.background_gradient(cmap="Reds")
```

## Visualising geographically spread

```python
df_spread = pd.merge(ind_coor,df,on="Name of State / UT")
df_spread
```

```python
map = folium.Map(location=[20,70],zoom_start=4,tiles = "Stamenterrain")
for lat,lon,value,name in zip(df_spread['Latitude'],df_spread['Longitude'],df_spread['Total Cases'],
    df_spread['Name of State / UT']): folium.CircleMarker([lat,lon],radius=value*0.8,color='red',fill_color='red',fill_opacity=0.3 ).add_to(map)
    
map
```

### Confirmed vs Recovered figures

```python
data = df_spread[['Name of State / UT','Total Cases','Cured','Death']]
data
```

```python
f,ax = plt.subplots(figsize = (15,8))
data.sort_values('Total Cases',ascending=False,inplace = True)
sns.set_color_codes("pastel")
sns.barplot(x = "Total Cases",y = "Name of State / UT",data = data,label = "Total",color = 'r')
sns.set_color_codes("muted")
sns.barplot(x = "Cured",y = "Name of State / UT",data = data,label = "Cured",color = 'g')
ax.legend(ncol = 2,loc = "lower right",frameon=True)
ax.set(xlim = (0,35),ylabel = "",xlabel = "Cases")
sns.despine(left = True,bottom = True)
```

```python
f,ax = plt.subplots(figsize = (15,8))
data.sort_values('Total Cases',ascending=False,inplace = True)
sns.set_color_codes("pastel")
sns.barplot(x = "Total Cases",y = "Name of State / UT",data = data,label = "Total",color = 'r')
sns.set_color_codes("muted")
sns.barplot(x = "Death",y = "Name of State / UT",data = data,label = "Death",color = 'b')
ax.legend(ncol = 2,loc = "lower right",frameon=True)
ax.set(xlim = (0,35),ylabel = "",xlabel = "Cases")
sns.despine(left = True,bottom = True)
```

##  Visualizing : Worldwide N COVID-19 cases

```python
confirmed = world_df.groupby('Date').sum()['Confirmed'].reset_index()
recovered = world_df.groupby('Date').sum()['Recovered'].reset_index()
death = world_df.groupby('Date').sum()['Deaths'].reset_index()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = confirmed['Date'] , y = confirmed['Confirmed'] , mode = "lines+markers" , name = "Confirmed" , 
                         line = dict(color = "Blue" , width = 2)))
fig.add_trace(go.Scatter(x = recovered['Date'] , y = recovered['Recovered'] , mode = "lines+markers" , name = "Recovered" , 
                         line = dict(color = "Green" , width = 2)))
fig.add_trace(go.Scatter(x = death['Date'] , y = death['Deaths'] , mode = "lines+markers" , name = "Deaths" , 
                         line = dict(color = "Red" , width = 2)))
fig.update_layout(title='Worldwide N COVID-19 Cases', xaxis_tickfont_size=14,xaxis=dict(title= 'Date'),
                  yaxis=dict(title='Number of Cases'))
fig.show()
```
